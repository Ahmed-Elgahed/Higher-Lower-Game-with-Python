```python
import random
from IPython.display import clear_output

data = [
    {'name': 'Instagram', 'follower_count': 346, 'description': 'Social media platform', 'country': 'United States'},
    {'name': 'Cristiano Ronaldo', 'follower_count': 215, 'description': 'Footballer', 'country': 'Portugal'},
    {'name': 'Ariana Grande', 'follower_count': 183, 'description': 'Musician and actress', 'country': 'United States'}
]

logo = """
  _    _ _       _                 _                            
 | |  | (_)     | |               | |                           
 | |__| |_  __ _| |__   ___ _ __  | |     _____      _____ _ __ 
 |  __  | |/ _` | '_ \ / _ \ '__| | |    / _ \ \ /\ / / _ \ '__|
 | |  | | | (_| | | | |  __/ |    | |___| (_) \ V  V /  __/ |   
 |_|  |_|_|\__, |_| |_|\___|_|    |______\___/ \_/\_/ \___|_|   
            __/ |                                               
           |___/                                                 
"""

vs = """
 _    __    
| |  / /____
| | / / ___/
| |/ (__  ) 
|___/____(_)
"""

def assign():
    return random.choice(data)

def compare(p1, p2, user_input):
    sum1 = p1["follower_count"]
    sum2 = p2["follower_count"]
    if sum1 > sum2:
        max_person = p1["name"]
    else:
        max_person = p2["name"]
    return max_person.lower() == user_input.lower()

def play_higher_lower():
    playing_game = True
    while playing_game:
        score = 0
        still_guessing = True
        person1 = assign()
        person2 = assign()
        while person1 == person2:
            person2 = assign()
        while still_guessing:
            clear_output(wait=True)
            print(logo)
            print(f"Compare A: {person1['name']}, {person1['description']}, from {person1['country']}")
            print(vs)
            print(f"Against B: {person2['name']}, {person2['description']}, from {person2['country']}")
            print("-----------------------------------")
            print(f"Your current score: {score}")
            print("-----------------------------------")
            guess = input("Enter name of person with higher followers: ")
            if compare(person1, person2, guess):
                score += 1
                person1 = person2
                person2 = assign()
                while person1 == person2:
                    person2 = assign()
            else:
                still_guessing = False
                print(f"Wrong! Final score: {score}")
        play_again = input("Want to play again? (y/n): ").lower()
        if play_again != "y":
            playing_game = False
            clear_output(wait=True)
            print("Game exited successfully")
```

```python
want_to_play = input("Do you want to play Higher-Lower? (y/n): ").lower()
if want_to_play == "y":
    play_higher_lower()
else:
    print("Program exit successful.")
```
